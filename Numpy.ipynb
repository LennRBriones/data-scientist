## Operaciones con arrays

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5, 6]
```

```python
array_1 = np.array(lista_1)
array_2 = np.array(lista_2)
```

```python
lista_1 + lista_2 #aquí sumamos las listas pero lo que hace es unirlas, por defecto 
                  #Python no te puede hacer sumas aritmeticas, pero numpy si con arrays
```

```python
array_1 + array_2 #Y este sería el meétodo
```

```python
array_1 * array_2
```

```python
#ahora lo mismo pero con el metodo arcaico y dificil, con list comprehensions
[ num1*num2 for num1, num2 in zip(lista_1, lista_2)]
```

# Estadistica

## Distribuciones

Aquí suponiendo que hacemos una ditribución normal, en la que la media de alturas en 1000 individuos es de 1,75mts, con una desviación estandar de 5cm
(el primer paramtero debe ser la media, el segundo la desviación y el tercero la cantidad de datos a generar)

```python
array_normal = np.random.normal(1.75, 0.05, 1000)
```

```python
array_normal
```

## np.mean

```python
np.mean(array_normal) #Con esto sacamos la media de una distribución
```

## np.median

```python
np.median(array_normal) #Sacar la mediana, que el es valor centro de una distribución
```

## np.std

```python
np.std(array_normal) #Sacar la desviación estandar si esta no fue dada
```
