# Construyendo una figura

```python
x = list(range(10))
y = x

fig = plt.figure(figsize = (5,5)) #Definir el tama√±o del cuadro

plt.plot(x, y, label= "Linea azulita") #definir el eje x, y  el nombre de la l√≠nea

plt.title("Una Linea") #Evidentemente el titulo

plt.xlabel("Eje de las x")  #Asignar distinci√≥n
plt.ylabel("Ejes de las y")

plt.legend() #Que me muestre la leyenda de como se llama la linea

plt.show() #Por convenci√≥n hacerlo para no afectar a los que lo usen en otro IDE
```

Podemos asignarlo a una funci√≥n para no tener que rehacer la l√≠nea

```python
x = list(range(10))
y = x

def generar_linea(eje_x, eje_y):
    fig = plt.figure(figsize = (5,5)) #Definir el tama√±o del cuadro
    
    plt.plot(eje_x, eje_y, label= "Linea azulita") #definir el eje x, y  el nombre de la l√≠nea
    
    plt.title("Una Linea") #Evidentemente el titulo
    
    plt.xlabel("Eje de las x")  #Asignar distinci√≥n
    plt.ylabel("Ejes de las y")
    
    plt.legend() #Que me muestre la leyenda de como se llama la linea
    
    plt.show() #Por convenci√≥n hacerlo para no afectar a los que lo usen en otro IDE
```

```python
generar_linea(x, y)
```

Tambi√©n podemos guardar la gr√°fica como imagen

```python
x = list(range(10))
y = x

fig = plt.figure(figsize = (5,5)) 

plt.plot(x, y, label= "Linea azulita") 

plt.title("Una Linea") 

plt.xlabel("Eje de las x")  
plt.ylabel("Ejes de las y")

plt.legend() 

plt.savefig("figura.png") #Aqu√≠ estar√°s guardando como imagen

plt.show() 
```

A su vez, puedes guardarla con fondo transparente, lo cual es guay

```python
x = list(range(10))
y = x

fig = plt.figure(figsize = (5,5)) 

plt.plot(x, y, label= "Linea azulita") 

plt.title("Una Linea") 

plt.xlabel("Eje de las x")  
plt.ylabel("Ejes de las y")

plt.legend() 

plt.savefig("figura_transparente.png", transparent= True) #Aqu√≠ estar√°s guardando como imagen

plt.show() 
```

Hay que cambiar el tipo de dato de fecha a datetime, pero... ¬øpor qu√©? lo averiguaremos con un ejemplo

```python
x = df["date"]
y = df["salary"]

fig = plt.figure(figsize=(18,18))

plt.plot(x, y)

plt.show()
```

Como podemos ver, es ilegible, al no tener la fecha como Datetime

De igual manera, podemos pedir cu√°ntos valores queremos que me traiga de los datos

```python
valores = 50

x = df["date"] [:valores] #Con esto indico de 0 a 10
y = df["salary"] [:valores]

fig = plt.figure(figsize=(18,18))

plt.plot(x, y)

plt.show()
```

La manera de arreglarlo es organizando los datos

```python
df["date"] = pd.to_datetime(df["date"]) #Los transformamos a datetime

df.sort_values(by="date", inplace=True)
```

```python

x = df["date"] 
y = df["salary"] 

fig = plt.figure(figsize=(18,18))

plt.plot(x, y)

plt.show()
```

```python
valores = 10
x = df["date"] [:valores] 
y = df["salary"] [:valores]

fig = plt.figure(figsize=(18,9))

plt.plot(x, y)

plt.show()
```

```python
# Definir fechas como indice
nuevo_df = df.set_index("date")

# Agruparlos por inicio de mes y calcular media
datos = nuevo_df.groupby(pd.Grouper(freq="MS")).agg(["mean"])
```

```python
plt.plot(datos)
```

```python
fig = plt.figure(figsize=(18,9))

plt.plot(datos)

plt.show()
```
